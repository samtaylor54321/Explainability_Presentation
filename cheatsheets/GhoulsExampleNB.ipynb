```python
train_df = pd.read_csv("../input/ghouls-goblins-and-ghosts-boo/train.csv")
test_df = pd.read_csv("../input/ghouls-goblins-and-ghosts-boo/test.csv")
```

```python
y_train = train_df["type"]
train_df.drop(["type"], axis=1, inplace=True)
```

```python
train_df.head()
```

```python
train_df = pd.get_dummies(train_df)
```

```python
train_df = train_df.drop('id',axis=1)
```

```python
X_train = train_df.values
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
```

```python
clf.fit(X_train, y_train.values)
```

```python
test_df = test_df.drop('id',axis=1)
test_df = pd.get_dummies(test_df)
test_df.head()
```

```python
y_pred = clf.predict(test_df.values)
```

```python
results = pd.read_csv("../input/ghouls-goblins-and-ghosts-boo/sample_submission.csv")
results["type"] = y_pred
results.to_csv("submission.csv", index=False)
```
